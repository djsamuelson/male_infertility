```python

def assign_sperm_quality_category(Sperm_quality):
    if Sperm_quality == 1:
        return 'Normal'
    elif Sperm_quality == 2:
        return 'Oligozoospermia'
    elif Sperm_quality == 3:
        return 'Asthenozoospermia'
    elif Sperm_quality == 4:
        return 'Teratozoospermia'
    elif Sperm_quality == 5:
        return 'Oligoasthenozoospermia'
    if Sperm_quality == 6:
        return 'Asthenoteratozoospermia'
    if Sperm_quality == 7:
        return 'Oligoasthenoteratozoospermia_OATs'
    if Sperm_quality == 8:
        return 'Azoospermia'
    else:
        return 'NaN'
    
df_fertility_stress['sperm_quality_category'] = df_fertility_stress['Sperm_quality'].apply(assign_sperm_quality_category)
df_fertility_stress[['sperm_quality_category']] = df_fertility_stress[['sperm_quality_category']].astype('category')

print(df_fertility_stress.shape)
df_fertility_stress.info()
#df_fertility_stress.unique
unique_values = df_fertility_stress['sperm_quality_category'].unique()
print("unique for sperm_quality_category:")
print(unique_values)
nunique_values = df_fertility_stress['sperm_quality_category'].nunique()
print("nunique for sperm_quality_category:")
print(nunique_values)
    
```

```python
unique_values = df_fertility_stress['Sperm_quality'].unique()
print("unique for Sperm_quality:")
print(unique_values)
nunique_values = df_fertility_stress['Sperm_quality'].nunique()
print("nunique for Sperm_quality:")
print(nunique_values)
print(df_fertility_stress['Sperm_quality'].count())
```

read csv files with visem molecular signature data and EDA using pandas

```python
df_semen_analysis = pd.read_csv('../data/visem_semen_analysis_data.csv', delimiter=';')
df_sex_hormones = pd.read_csv('../data/visem_sex_hormones.csv', delimiter=';')
df_subject_metrics = pd.read_csv('../data/visem_participant_related_data.csv', delimiter=';')
df_fa_serum = pd.read_csv('../data/visem_fatty_acids_serum.csv', delimiter=';')
df_fa_sperm = pd.read_csv('../data/visem_fatty_acids_spermatoza.csv', delimiter=';')
```

merge male visem fertility molecular signature datasets on 'ID'

```python
merge_df1 = pd.merge(df_subject_metrics, df_semen_analysis, on ='ID', how = 'inner')
merge_df2 = pd.merge(merge_df1, df_sex_hormones, on ='ID', how = 'inner')
merge_df3 = pd.merge(merge_df2, df_fa_serum, on ='ID', how = 'inner')
df_mol_sig = pd.merge(merge_df3, df_fa_sperm, on ='ID', how = 'inner')
```
